```python
state_testing_csv = pd.read_csv('http://covidtracking.com/api/states/daily.csv')
counties_nyt_csv = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv')
```

```python
counties_nyt = df(counties_nyt_csv)
```

```python
counties_length = len(counties_nyt)
counties_nyt
```

```python
#county = 'Goodhue'
county = 'Manatee'
county_history = []
for i in range(counties_length):
    if counties_nyt.county[i] == county:
        county_slice = counties_nyt[i:i+1]
        #Dates in .csv are already chrono, don't need to double check county_slice.date
        county_history.append([county_slice.cases[i],county_slice.deaths[i]])
```

```python
county_length = len(county_history)
county_length, county_history
```

```python
county_history_matrix = np.array(county_history)
```
